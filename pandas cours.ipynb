```python
#Concaténation (Verticale et Horizontale) :

# DataFrames d'exemple
df1 = pd.DataFrame({'A': [1, 2], 'B': [3, 4]})
df2 = pd.DataFrame({'A': [5, 6], 'B': [7, 8]})

# Concaténation verticale
result_vertical = pd.concat([df1, df2], axis=0)
print(result_vertical)
df1 = pd.DataFrame({'A': [1, 2], 'B': [3, 4]})
df2 = pd.DataFrame({'f': [5, 6], 'r': [7, 8]})
result_vertical = pd.concat([df1, df2], axis=0)
print(result_vertical)
# Concaténation horizontale
result_horizontal = pd.concat([df1, df2], axis=1)
print(result_horizontal)
```

```python
# Exemple de DataFrame
df = pd.DataFrame({'Date': ['2021-01-01', '2021-02-01', '2021-03-01'], 'Valeur': [10, 20, 30]})

# Conversion de la colonne 'Date' en format datetime
df["Date"] = pd.to_datetime(df["Date"])

# Extraction de l'année
df["Année"] = df["Date"].dt.year

# Extraction du mois
df["Mois"] = df["Date"].dt.month

# Extraction du jour
df["Jour"] = df["Date"].dt.day

print(df)
#filtrage:
df_filtered = df[(df["Date"] >= "2024-01-01") & (df["Date"] <= "2024-12-31")]
df_2024 = df[df["Date"].dt.year == 2024]
df_janvier = df[df["Date"].dt.month == 1]
```

```python
# Resampling : Agréger les données par mois
df = pd.DataFrame({
    'Date': ['2021-01-01', '2021-01-15', '2021-02-01', '2021-02-15'],
    'Valeur': [10, 20, 30, 40]
})

# Conversion de la colonne 'Date' en format datetime
df["Date"] = pd.to_datetime(df["Date"])
df_resampled = df.resample("M", on="Date").sum()
print(df_resampled)
```

```python
# Création d'un DataFrame
df = pd.DataFrame({
    'Valeurs': [1, 2, 3, 4, 5]
})

# Appliquer une fonction lambda pour ajouter 10 à chaque élément
df['Valeurs'] = df['Valeurs'].apply(lambda x: x + 10) #axis=0 by default

print(df)
# Fonction personnalisée
def au_carre(x):
    return x ** 2

# Appliquer la fonction sur la colonne 'Valeurs'
df['Valeurs'] = df['Valeurs'].apply(au_carre)
print(df)

# Création d'un DataFrame
df = pd.DataFrame({
    'A': [1, 2, 3],
    'B': [4, 5, 6]
})

# Appliquer une fonction ligne par ligne pour calculer la somme
df['Total'] = df.apply(lambda row: row['A'] + row['B'], axis=1)

print(df)
# Fonction personnalisée
def multiplication_ligne(row):
    return row['A'] * row['B']

# Appliquer la fonction personnalisée ligne par ligne
df['Produit'] = df.apply(multiplication_ligne, axis=1)

print(df)
```

```python
import numpy as np
df=pd.DataFrame({"couleur":[1,66,3,45,4,44,0],"age":[18,19,None,20,15,33,4]})
print(df)
print(df.loc[1]["age"])
df = df.applymap(lambda x: x + 10)
print(df)
#une seule condition sur colone
df['condition one']=df['age'].apply(lambda x : "eleve" if x>30 else "faible")
print(df)
# Définir les conditions et les résultats correspondants plusieurs conditions sur meme colone
df=pd.DataFrame({"couleur":[1,66,3,45,4,44,0],"age":[18,19,None,20,15,33,4]})

conditions = [df['age'] < 18 , (df['age'] >= 18) & (df['age'] <= 40)  , df['age'] > 40]

valeurs = ['Enfant', 'Adulte', 'Senior']

# Créer une nouvelle colonne avec np.select()
df['Catégorie'] = np.select(conditions, valeurs)

print(df)
```

```python
# DataFrame avec des valeurs répétitives
df = pd.DataFrame({
    "Colonne": ["Rouge", "Bleu", "Vert", "Rouge", "Vert"]
})

# Conversion en type catégorique ???????????
df["Colonne"] = df["Colonne"].astype("category")

print(df)
print(df.dtypes)
print(df["Colonne"].cat.codes)  # Codes internes des catégories
```

Qu'est-ce que le type catégorique=modalites ?

C'est un type spécial pour des colonnes avec des valeurs répétitives (catégories).
Exemple ici : les valeurs "Rouge", "Bleu", et "Vert" sont les catégories.
Pourquoi utiliser le type catégorique ?

Réduction de la mémoire utilisée (utile pour des données volumineuses).
Accélération des opérations sur la colonne.
Que fait astype("category") ?

Remplace chaque valeur par un code numérique interne :
"Bleu" → 0
"Vert" → 1
"Rouge" → 2
Ces codes sont accessibles avec df["Colonne"].cat.codes
En résumé, le type catégorique est idéal pour des colonnes avec un nombre limité de valeurs distinctes (comme des couleurs, des villes, ou des types).
